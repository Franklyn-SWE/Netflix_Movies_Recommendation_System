```python
#Feature Engineering (selection )

movies.columns
```

```python
movies=movies[['id', 'title', 'overview','genre' ]] 
movies
```

```python
# Tagging genre and overview of the movies
```

```python
movies['tags'] = movies['overview']+movies['genre']
movies
```

```python
# dropping overview and genre columns 
```

```python
new_data = movies.drop(columns=['overview', 'genre'])
```

```python
new_data
```

```python
# coverting to vector by importing the vector count  features 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=1000, stop_words='english')
cv
```

```python

cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
# Viewing the vector shape to confirm 10,000 columns and 10,000 rows
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
vector.shape
```

```python
# Using Cosine similarity for comparison and recommendtion 
# max_feature = 10,000
# hence, counvector = 10,000 
```

```python
# importing cosine similarity and implementation 
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity
```

```python
# Using the index to identify movies 
```

```python
# checking for the movie title named The Godfather as an example
new_data[new_data['title']=="The Godfather"].index[0]
```

```python
similarity[2]
```

```python
sorted(list(enumerate(similarity[2])), reverse=True)
```

```python
# value will be access using vector
sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
```

```python
# Viewing the info above shows the datatypes, hence the titles are needed
```

```python
# checking the info 
new_data.info()
```

```python
# printing the movies similarity based on the titles ( using for loop to list the first 5 movies)
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for x in distance[1:6]:
    print(new_data.iloc[x[0]].title)
```

```python
# Creating a function to recommens movies in general:
def recommends(movies):
    index=new_data[new_data['title']==movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    for x in distance[1:6]:
        print(new_data.iloc[x[0]].title)
    
```

```python
# calling the functions by inserting a movie tile for recommendation 
recommends("Iron Man")
```
